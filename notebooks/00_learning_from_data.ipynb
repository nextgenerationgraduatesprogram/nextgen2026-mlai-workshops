# 00 - Learning From Data

## Scientific hypotheses, observations, and learning

Learning can be read as a computational version of comparing explanations against evidence.


A scientific hypothesis is a claim about how part of the world behaves. It links observations to expectations. Evidence rarely gives direct access to truth; it narrows the explanations we can defend.

Machine learning makes the same loop computational. We decide what explanations are allowed, collect evidence, and define rules for comparing those explanations. The learned model is the result of those choices applied to data.

## 3. The Learning Problem

The workshop frame separates possible explanations, evidence, selection rules, and the selected solution.


The workshop uses one recurring frame:

$$
\boxed{\mathcal{H} + \mathcal{D} + \mathcal{O} \longrightarrow s}
$$

- $\mathcal{H}$ is the **hypothesis space**: the functions the learner may consider.
- $\mathcal{D}$ is the **dataset**: the evidence available.
- $\mathcal{O}$ is the **optimization space**: losses, penalties, priors, algorithms, validation protocols, and selection rules.
- $s$ is the **selected solution**: the fitted rule, parameter vector, posterior summary, or decision procedure.
- $\mathcal{S}$ is the **acceptable solution space**: the set of solutions we would trust for the problem.

The frame is a diagnostic. When a model behaves well or badly, ask which part shaped that behaviour: what was possible, what evidence was available, what rule selected $s$, and whether $s\in\mathcal{S}$.

## 5. The Running Problem

The solar-panel tilt example keeps one simple input-output problem visible across the workshop.


Every notebook uses the same calibration example. We set a solar panel to a normalized tilt $x$ and observe a noisy power output $y$. The goal is to learn a rule that predicts how power output changes as the panel is tilted.

The panel system has a hidden response rule:

$$
f^*: \mathcal{X} \rightarrow \mathcal{Y}.
$$

We do not observe $f^*$ directly. We observe finite noisy readings:

$$
y_i = f^*(x_i) + \epsilon_i.
$$

Learning means selecting an approximate rule $h$ from a hypothesis space $\mathcal{H}$. The selected rule is not the panel system; it is a defensible approximation from limited calibration evidence.

## 6. Concrete Running Story

The running story defines the variables used in every notebook.


A useful way to read the example is as a solar-panel calibration problem.

- $x$ is the solar-panel tilt angle, measured as a fraction of one allowed tilt range from 0 to 1.
- $y$ is the observed solar-power output, represented in normalized units so the same regression ideas stay visible.
- $f^*(x)$ is the hidden power-output curve we would like to understand.
- $\epsilon_i$ is sensor noise, irradiance fluctuation, timing variation, or other unrecorded context in the measurement process.

The notebooks use a simulated hidden power-output curve so we can see what the learner cannot normally see. In a real calibration task, the true response curve is unavailable, so we rely on data, assumptions, validation, and domain judgement.

## 7. What Machine Learning Adds

Machine learning scales model comparison when there are too many candidate explanations to inspect by hand.


Machine learning gives us computational machinery for scientific modelling when the set of possible explanations is too large to compare by hand. We define what kinds of explanations are allowed, collect finite evidence, and use selection machinery to choose one result from many candidates.

That matters because a learned model can help us predict power output where direct measurement is expensive, compress many observations into a useful pattern, expose where current evidence is weak, and decide where another measurement might be valuable.

The machinery will get richer across the workshop, but the central diagnostic question stays the same:

> What hypotheses were available, what data shaped the result, what optimization choices selected $s$, and is $s$ inside the solution space $\mathcal{S}$ we would accept for this problem?


### 7.1 Probability side note: Learning as Updating Belief

One probabilistic way to read the same frame is Bayes' rule:

$$
p(h\mid \mathcal{D}) \propto p(\mathcal{D}\mid h)p(h).
$$

Here $p(h)$ describes which hypotheses were plausible before the dataset, $p(\mathcal{D}\mid h)$ describes how compatible the observed measurements are with a hypothesis, and $p(h\mid \mathcal{D})$ describes updated plausibility after observing the data.

The notebooks mostly use optimization language, but the probability view is nearby throughout. Likelihoods score compatibility with data. Penalties and priors express preferences among solutions. Validation asks whether the selected solution deserves trust beyond the evidence that selected it.
