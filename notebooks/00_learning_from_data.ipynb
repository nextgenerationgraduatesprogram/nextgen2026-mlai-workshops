## Scientific hypotheses, observations, and learning

A scientific hypothesis is a theory about how some part of the world behaves. It connects observations to expectations. A useful hypothesis should help explain observations we have already made and support predictions about observations we have yet to make. Evidence rarely gives direct access to truth. It constrains which explanations remain plausible. 

Machine learning begins from the same position, then makes the loop computational: we specify what kinds of explanations are allowed, collect evidence, and define rules that compare or summarise the allowed explanations. The learned model is the result of those modelling choices acting on the data.

## The Learning Problem

This workshop studies machine learning through one recurring decomposition:

$$
\boxed{\mathcal{H} + \mathcal{D} + \mathcal{O} \longrightarrow s}
$$

The symbols are a compact way to name the ingredients of a learning problem.

- $\mathcal{H}$ is the **hypothesis space**: the explanations or functions the learner is allowed to consider.
- $\mathcal{D}$ is the **dataset**: the evidence available to constrain those explanations.
- $\mathcal{O}$ is the **optimization space**: the losses, penalties, priors, algorithms, validation protocols, and selection rules used to compare possibilities. In this workshop, $\mathcal{O}$ is broader than a numerical optimizer; it names the whole selection machinery.
- $s$ is the **selected solution**: the fitted model, parameters, posterior summary, prediction rule, or decision procedure produced by the learning process.
- $\mathcal{S}$ is the **solution space**: the set of solutions we would consider acceptable for the problem. Ideally, the learning process returns $s \in \mathcal{S}$, but that is something we have to check rather than assume.

Use this box as a diagnostic lens. When a model behaves well or badly, we can ask which part of the recipe shaped that behaviour: what was possible, what evidence was available, what rule selected $s$, and whether $s$ belongs in the solution space we actually care about.

## The Running Problem

To keep the workshop connected, every notebook uses the same simple problem: we observe a noisy response $y$ at input locations $x$, and we want to learn a function that predicts or explains how the response changes with the input. 

Think of the world as having some underlying response rule:

$$
f^*: \mathcal{X} \rightarrow \mathcal{Y}.
$$

We do not observe $f^*$ directly. We observe finite, noisy measurements from a process such as

$$
y_i = f^*(x_i) + \epsilon_i.
$$

The learning task is to use those measurements to select an approximate rule $h$ from some hypothesis space $\mathcal{H}$. The selected rule will not be the world itself; it is a defensible approximation built from limited evidence.

The specific story is intentionally generic so the modelling choices stay visible. In practice, the hidden signal is usually unavailable, so we judge a fitted rule through observed evidence, held-out checks, assumptions, and the intended use.

## Concrete Running Story

A useful way to read the example is as an instrument-calibration problem.

- $x$ is a controlled input, such as dose, time, temperature, concentration, or simulation setting.
- $y$ is a measured response, such as signal intensity, growth, error, yield, or score.
- $f^*(x)$ is the hidden response rule we would like to understand.
- $\epsilon_i$ is measurement noise, biological variation, simulation randomness, annotation disagreement, or unrecorded context.

The notebooks use a simulated hidden rule so we can see what the learner cannot normally see. In real research, the hidden rule is unavailable, so we rely on data, assumptions, validation, and domain judgement.


## What Machine Learning Adds

Machine learning gives us computational machinery for scientific modelling when the set of possible explanations is too large to compare by hand. We define what kinds of explanations are allowed, collect finite evidence, and use selection machinery to choose one result from many candidates.

That matters because a learned model can help us predict where direct measurement is expensive, compress many observations into a useful pattern, expose where current evidence is weak, and decide where another measurement might be valuable.

The machinery will get richer across the workshop, but the central diagnostic question stays the same:

> What hypotheses were available, what data shaped the result, what optimization choices selected $s$, and is $s$ inside the solution space $\mathcal{S}$ we would accept for this problem?


### Probability side note: Learning as Updating Belief

One probabilistic way to read the same frame is Bayes' rule:

$$
p(h\mid \mathcal{D}) \propto p(\mathcal{D}\mid h)p(h).
$$

Here $p(h)$ describes which hypotheses were plausible before the dataset, $p(\mathcal{D}\mid h)$ describes how compatible the observed data are with a hypothesis, and $p(h\mid \mathcal{D})$ describes updated plausibility after observing the data.

The notebooks mostly use optimization language, but the probability view is nearby throughout. Likelihoods score compatibility with data. Penalties and priors express preferences among solutions. Validation asks whether the selected solution deserves trust beyond the evidence that selected it.
