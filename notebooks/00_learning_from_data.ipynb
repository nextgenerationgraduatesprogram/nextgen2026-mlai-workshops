# 00 - Learning From Data

## Scientific hypotheses, observations, and learning

A scientific hypothesis is a claim about how part of the world behaves. It links observations to expectations. Evidence rarely gives direct access to truth; it narrows the explanations we can defend.

Machine learning makes the same loop computational. We decide what relationships are allowed, collect evidence about those relationships, and define rules for selecting one useful approximation. The learned model is the result of those choices applied to data.

## 2. Why Learning Functions Matters

Many machine learning problems begin with the same need: we want an answer for a case we have not directly observed.


If we already knew the exact rule, we would not need to learn it. If we could measure every possible case, we could look up the answer. Learning becomes useful when the relationship matters, direct measurement is limited, and we need a reusable approximation.

A **function** is a rule that turns what we know into what we want to know:

$$
f: \mathcal{X} \rightarrow \mathcal{Y}.
$$

In this notation, $x\in\mathcal{X}$ is the information available at prediction time and $y\in\mathcal{Y}$ is the output we want. Learning a function means using examples to build an approximate rule $h$ that can be used on new inputs.

### 2.1 Concrete example: solar-panel calibration

Suppose we are calibrating a solar panel. We can set the panel to a tilt $x$ and measure the power output $y$.

The direct measurement question is:

> What power did we observe at the tilts we tried?

The learning question is different:

> Given a tilt we may not have measured before, what power output should we expect?

That second question asks for a function. We want a rule that maps panel tilt to expected power output, because we cannot measure every possible tilt under every possible condition. A useful learned rule lets us compare settings, choose operating points, and identify where another measurement would be valuable.

In the real system there may be a hidden response function $f^*(x)$, but we do not get to see it directly. We only get noisy measurements:

$$
y_i = f^*(x_i) + \epsilon_i.
$$

The noise term $\epsilon_i$ stands for sensor error, irradiance changes, timing differences, and other context we did not record. The data points are evidence about the function, not the function itself.

So the learner's job is not to recover the true physical system perfectly. The job is to choose an approximation $h$ that is good enough for the decision we need to make:

$$
h(x) \approx f^*(x).
$$

### 2.2 Back to the general learning problem

The solar-panel example is a simple version of a general supervised learning problem. We start with examples,

$$
\mathcal{D}=\{(x_i,y_i)\}_{i=1}^n,
$$

and use them to select a function $h$ that will be useful on future inputs. The same pattern appears when pixels map to labels, symptoms map to risk estimates, prompts map to responses, or operating conditions map to system behaviour.

This is why function approximation sits at the centre of the learning problem. The dataset $\mathcal{D}$ gives finite evidence. The model class decides which functions are even possible. The selection rule decides which possible function is chosen. The result is a learned approximation, not a guarantee of truth.

The function view gives us the diagnostic questions for the rest of the workshop:

1. What information is available as input $x$?
2. What output $y$ do we need for the intended decision?
3. What examples in $\mathcal{D}$ constrain the relationship between them?
4. What kinds of functions are possible before fitting begins?
5. What rule selects one approximation, and where must it be trusted?

## 3. The Learning Problem

The workshop frame names the pieces of the function-learning problem.


Section 2 described learning as choosing an approximate function $h$ from examples. The workshop uses one recurring frame to keep that choice inspectable:

$$
\boxed{\mathcal{H} + \mathcal{D} + \mathcal{O} \longrightarrow s}
$$

- $\mathcal{H}$ is the **hypothesis space**: the functions the learner may consider.
- $\mathcal{D}$ is the **dataset**: the examples and measurements that constrain those functions.
- $\mathcal{O}$ is the **optimization space**: losses, penalties, priors, algorithms, validation protocols, and selection rules.
- $s$ is the **selected solution**: the fitted function, parameter vector, posterior summary, or decision procedure.
- $\mathcal{S}$ is the **acceptable solution space**: the set of solutions we would trust for the intended use.

For the solar-panel problem, $\mathcal{D}$ contains measured tilt-output pairs, $\mathcal{H}$ defines the kinds of tilt-power functions that are available, $\mathcal{O}$ decides which available function is selected, and $s$ is the learned calibration rule. The frame is a diagnostic. When a model behaves well or badly, ask which part shaped that behaviour: what was possible, what evidence was available, what rule selected $s$, and whether $s\in\mathcal{S}$.

## 4. The Running Example

The solar-panel calibration problem gives the abstract frame a concrete mapping.


Every notebook uses the same simple input-output problem. We set a solar panel to a normalized tilt $x$ and observe a noisy power output $y$. The goal is to learn a function that predicts power output at tilts we have not directly measured.

Read the variables this way:

- $x$ is the solar-panel tilt angle, measured as a fraction of one allowed tilt range from 0 to 1.
- $y$ is the observed solar-power output, represented in normalized units so the same regression ideas stay visible.
- $f^*(x)$ is the hidden power-output relationship we would like to understand but cannot observe directly.
- $h(x)$ is the learned approximation we can actually use for prediction and decision-making.
- $\epsilon_i$ is sensor noise, irradiance fluctuation, timing variation, or other unrecorded context in the measurement process.

The hidden response rule is

$$
f^*: \mathcal{X} \rightarrow \mathcal{Y}.
$$

We do not observe $f^*$ directly. We observe finite noisy readings:

$$
y_i = f^*(x_i) + \epsilon_i.
$$

Learning means selecting an approximate rule $h$ from a hypothesis space $\mathcal{H}$. The selected rule is not the panel system itself; it is a defensible approximation from limited calibration evidence. That approximation is useful only if it supports the intended decision: comparing tilts, choosing settings, or deciding where another measurement is needed.

The notebooks use a simulated hidden power-output curve so we can see what the learner cannot normally see. In a real calibration task, the true response curve is unavailable, so we rely on data, assumptions, validation, and domain judgement.
